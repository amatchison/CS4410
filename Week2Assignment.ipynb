```python
%matplotlib inline
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()  # Bunch object
```

```python
#display dataset's description
print(diabetes.DESCR)
```

```python
#Confirm number of samples/features
diabetes.data.shape
```

```python
#Confirm number of targets
diabetes.target.shape
```

```python
#Confirm feature names
diabetes.feature_names
```

```python
import pandas as pd
```

```python
pd.set_option('display.precision', 4)  # 4 digit precision for floats
```

```python
# Used for command line outputs in IPython interactive mode
pd.set_option('display.max_columns', 11)  # display up to 9 columns in DataFrame outputs

pd.set_option('display.width', None)  # auto-detect the display width for wrapping
```

```python
#create dataframe
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
#create column for DiseaseProgression
diabetes_df['DiseaseProgression'] = pd.Series(diabetes.target)
```

```python
diabetes_df.head()  # peek at first 5 rows
```

```python
#calculate summary statistics
diabetes_df.describe()
```

```python
#create sample
sample_df = diabetes_df.sample(frac=0.1, random_state=17)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#set font size
sns.set(font_scale=1)
```

```python
#set  grid style
sns.set_style('whitegrid')
```

```python
#display scatterplots for each feature
for feature in diabetes.feature_names:
    plt.figure(figsize=(8, 4.5))  # 8"-by-4.5" Figure
    sns.scatterplot(data=sample_df, x=feature,
                    y='DiseaseProgression', hue='DiseaseProgression',
                    palette='cool', legend=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split datat into train and test
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data, diabetes.target, random_state=11)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression = LinearRegression()
```

```python
# training the model
linear_regression.fit(X=X_train, y=y_train)
```

```python
for i, name in enumerate(diabetes.feature_names):
    print(f'{name:>10}: {linear_regression.coef_[i]}')
```

```python
linear_regression.intercept_
```

```python
predicted = linear_regression.predict(X_test)
```

```python
expected = y_test
```

```python
predicted[:5]  # first 5 predictions
```

```python
expected[:5]   # first five targets
```

```python
# create dataframe for expected and predicted values
df = pd.DataFrame()
```

```python
df['Expected'] = pd.Series(expected)
```

```python
df['Predicted'] = pd.Series(predicted)
```

```python
# Display scatter plot with the expected values along the x-axis and the predicted values along the y-axis:

figure = plt.figure(figsize=(9, 9))

axes = sns.scatterplot(data=df, x='Expected', y='Predicted',
    hue='Predicted', palette='cool', legend=False)

start = min(expected.min(), predicted.min())

end = max(expected.max(), predicted.max())

axes.set_xlim(start, end)

axes.set_ylim(start, end)

line = plt.plot([start, end], [start, end], 'k--')
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(expected, predicted)
```

```python
from sklearn.linear_model import ElasticNet, Lasso, Ridge
```

```python
# check other estimators
estimators = {
    'LinearRegression': linear_regression,
    'ElasticNet': ElasticNet(),
    'Lasso': Lasso(),
    'Ridge': Ridge()
}
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
for estimator_name, estimator_object in estimators.items():
    kfold = KFold(n_splits=10, random_state=11, shuffle=True)
    scores = cross_val_score(estimator=estimator_object,
        X=diabetes.data, y=diabetes.target, cv=kfold,
        scoring='r2')
    print(f'{estimator_name:>16}: ' +
          f'mean of r2 scores={scores.mean():.3f}')
```

Linear Regression appears to be the best estimator.
